```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
import numpy as np
import constants


class DataSimulator():
    def __init__(self, random_seed=5170):
        self.random_seed = random_seed
        # Sets filter probability for each category
        self.filter_probs= {
            1: [0.1, 0.2, 0.3, 0.4],
            2: [0.4, 0.1, 0.2, 0.3],
            3: [0.3, 0.4, 0.1, 0.2],
            4: [0.2, 0.3, 0.4, 0.1],
            5: [0.2, 0.4, 0.1, 0.3]
        }
        self.config_probs = (0.4, 0.2, 0.1, 0.3)

    def generate_samples(self, n_samples=50):
        np.random.seed(self.random_seed)
        samples = []
        category_ids = np.random.choice(list(constants.CATEGORIES.keys()), size=n_samples)
        for id in category_ids:
            filter_id = np.random.choice(list(constants.FILTERS.keys()),
                                         p=self.filter_probs[id])
            config_id = np.random.choice(list(constants.FILTER_CONFIGS.keys()),
                                         p=self.config_probs)
            samples.append((id, filter_id, config_id))
        return samples

sim = DataSimulator()
samples = sim.generate_samples(500)
```

```python
categories = []
f1, f2, f3, f4, f5 = [], [], [], [], []
c1, c2, c3, c4, c5 = [], [], [], [], []

for i in range(len(samples)):
    categories.append(samples[i][0])
    if samples[i][0] == 1:
        f1.append(samples[i][1])
        c1.append(samples[i][2])
    elif samples[i][0] == 2:
        f2.append(samples[i][1])
        c2.append(samples[i][2])
    elif samples[i][0] == 3:
        f3.append(samples[i][1])
        c3.append(samples[i][2])
    elif samples[i][0] == 4:
        f4.append(samples[i][1])
        c4.append(samples[i][2])
    elif samples[i][0] == 5:
        f5.append(samples[i][1])
        c5.append(samples[i][2])

sns.displot(categories)

fig, ax = plt.subplots(5,2, figsize=(10, 10))
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)

ax[0][0].title.set_text(list(constants.CATEGORIES.values())[0])
sns.histplot(f1, ax=ax[0][0])
sns.histplot(c1, ax=ax[0][1])

ax[1][0].title.set_text(list(constants.CATEGORIES.values())[1])
sns.histplot(f2, ax=ax[1][0])
sns.histplot(c2, ax=ax[1][1])

ax[2][0].title.set_text(list(constants.CATEGORIES.values())[2])
sns.histplot(f3, ax=ax[2][0])
sns.histplot(c3, ax=ax[2][1])

ax[3][0].title.set_text(list(constants.CATEGORIES.values())[3])
sns.histplot(f4, ax=ax[3][0])
sns.histplot(c4, ax=ax[3][1])

ax[4][0].title.set_text(list(constants.CATEGORIES.values())[4])
sns.histplot(f5, ax=ax[4][0])
sns.histplot(c5, ax=ax[4][1])
```
